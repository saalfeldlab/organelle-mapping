```python
import yaml
from pathlib import Path
import fibsem_tools as fst
```

```python
with open("../all_data.yaml", "r") as f:
    all_data = yaml.safe_load(f)
    for data_name, data in all_data["datasets"].items():
        for crops in data["crops"]:
            for crop in crops.split(","):
                fzarr = fst.access(Path(all_data["gt_path"]) / data_name / "groundtruth.zarr" / crop, "r+")
                for label in fzarr.keys():
                    attrs = fzarr[label].attrs.asdict()
                    try:
                        n = attrs["multiscales"][0]
                    except KeyError:
                        attrs = attrs["multiscales"]
                        print("yes")
                    attrs["multiscales"][0]["name"] = ""
                    fzarr[label].attrs["multiscales"] = attrs["multiscales"]
```
